```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('./airplane/train.csv')
test_df = pd.read_csv('./airplane/test.csv')
```

## Data preprocessing

```python
# 결측치 제거
train_df = train_df.dropna()
test_df = test_df.dropna()
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

train_df['Gender'] = label_encoder.fit_transform(train_df['Gender'])
train_df['Customer Type'] = label_encoder.fit_transform(train_df['Customer Type'])
train_df['Type of Travel'] = label_encoder.fit_transform(train_df['Type of Travel'])
train_df['Class'] = label_encoder.fit_transform(train_df['Class'])
train_df['satisfaction'] = label_encoder.fit_transform(train_df['satisfaction'])
```

```python
test_df['Gender'] = label_encoder.fit_transform(test_df['Gender'])
test_df['Customer Type'] = label_encoder.fit_transform(test_df['Customer Type'])
test_df['Type of Travel'] = label_encoder.fit_transform(test_df['Type of Travel'])
test_df['Class'] = label_encoder.fit_transform(test_df['Class'])
test_df['satisfaction'] = label_encoder.fit_transform(test_df['satisfaction'])
```

```python
corr_matrix = train_df.corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
def generate_colors(num):
    if num==1:
        colors=''
        lst = list('ABCDEF0123456789')
        for i in range(num):
            colors='#'+''.join(np.random.choice(lst, 6))
    else:
        colors = []
        lst = list('ABCDEF0123456789')

        for i in range(num):
            colors.append('#'+''.join(np.random.choice(lst, 6)))
        
    return colors

detail_satisfaction_columns = ['Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking',
                                'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
                                'Inflight entertainment', 'On-board service', 'Leg room service', 'Baggage handling',
                                'Checkin service', 'Inflight service', 'Cleanliness']

fig, axes = plt.subplots(nrows=7, ncols=2, figsize=(10, 12))

for i,column_name in enumerate(detail_satisfaction_columns):
    row = i // 2
    col = i % 2
    ax = sns.boxplot(x=column_name, data=train_df, color=generate_colors(1), ax=axes[row, col])
    ax.set_title(column_name, size=6)
    ax.set_xlabel('')
    
plt.tight_layout()
plt.suptitle("Distribution of Numerical Columns", size=12, fontweight='bold', y=0)
plt.show()
```

## Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 특성과 레이블 설정
X_train = train_df.drop(['satisfaction'], axis=1)
y_train = train_df['satisfaction']

X_test = test_df.drop(['satisfaction'], axis=1)
y_test = test_df['satisfaction']

# 데이터 분할
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Random Forest 모델 생성
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 모델 학습
model.fit(X_train, y_train)

# 검증 데이터로 예측
y_val_pred = model.predict(X_val)

# 검증 데이터 평가
accuracy = accuracy_score(y_val, y_val_pred)
conf_matrix = confusion_matrix(y_val, y_val_pred)
classification_rep = classification_report(y_val, y_val_pred)

print(f'Validation Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')

# 테스트 데이터 예측
y_test_pred = model.predict(X_test)

# 테스트 데이터 평가
test_accuracy = accuracy_score(y_test, y_test_pred)
test_conf_matrix = confusion_matrix(y_test, y_test_pred)
test_classification_rep = classification_report(y_test, y_test_pred)

print(f'Test Accuracy: {test_accuracy}')
print(f'Test Confusion Matrix:\n{test_conf_matrix}')
print(f'Test Classification Report:\n{test_classification_rep}')
```

```python
conf_matrix = confusion_matrix(y_test, y_test_pred)

# 혼동 행렬을 정규화
conf_matrix_normalized = conf_matrix / conf_matrix.sum(axis=1)[:, np.newaxis]

# seaborn을 사용하여 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_normalized, annot=True, fmt=".4f", cmap="Blues", xticklabels=['Not Satisfied', 'Satisfied'], yticklabels=['Not Satisfied', 'Satisfied'])
plt.title('Normalized Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Get feature importances
feature_importance_val = model.feature_importances_

# Create a DataFrame with feature names and their importance scores for validation set
feature_importance_val_df = pd.DataFrame({
    'Feature': ['Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class',
                'Flight Distance', 'Inflight wifi service',
                'Departure/Arrival time convenient', 'Ease of Online booking',
                'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
                'Inflight entertainment', 'On-board service', 'Leg room service',
                'Baggage handling', 'Checkin service', 'Inflight service',
                'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'Generation'],
    'Importance': feature_importance_val
})

# Sort the DataFrame by importance in descending order
feature_importance_val_df = feature_importance_val_df.sort_values(by='Importance', ascending=False)

# Optionally, you can plot a bar chart for better visualization
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_val_df, palette='viridis')
plt.title('Random Forest Feature Importance on Validation Set')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Gradient Boosting 모델 생성
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)

# 모델 학습
gb_model.fit(X_train, y_train)

# 검증 데이터로 예측
y_val_pred_gb = gb_model.predict(X_val)

# 검증 데이터 평가
accuracy_gb = accuracy_score(y_val, y_val_pred_gb)
conf_matrix_gb = confusion_matrix(y_val, y_val_pred_gb)
classification_rep_gb = classification_report(y_val, y_val_pred_gb)

print(f'Gradient Boosting Validation Accuracy: {accuracy_gb}')
print(f'Gradient Boosting Confusion Matrix:\n{conf_matrix_gb}')
print(f'Gradient Boosting Classification Report:\n{classification_rep_gb}')

# 테스트 데이터로 예측
y_test_pred_gb = gb_model.predict(X_test)

# 테스트 데이터 평가
test_accuracy_gb = accuracy_score(y_test, y_test_pred_gb)
test_conf_matrix_gb = confusion_matrix(y_test, y_test_pred_gb)
test_classification_rep_gb = classification_report(y_test, y_test_pred_gb)

print(f'Gradient Boosting Test Accuracy: {test_accuracy_gb}')
print(f'Gradient Boosting Test Confusion Matrix:\n{test_conf_matrix_gb}')
print(f'Gradient Boosting Test Classification Report:\n{test_classification_rep_gb}')
```

```python
conf_matrix_normalized = test_conf_matrix_gb / test_conf_matrix_gb.sum(axis=1)[:, np.newaxis]

# seaborn을 사용하여 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_normalized, annot=True, fmt=".4f", cmap="Blues", xticklabels=['Not Satisfied', 'Satisfied'], yticklabels=['Not Satisfied', 'Satisfied'])
plt.title('Normalized Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Get feature importances
feature_importance_val = gb_model.feature_importances_

# Create a DataFrame with feature names and their importance scores for validation set
feature_importance_val_df = pd.DataFrame({
    'Feature': ['Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class',
                'Flight Distance', 'Inflight wifi service',
                'Departure/Arrival time convenient', 'Ease of Online booking',
                'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
                'Inflight entertainment', 'On-board service', 'Leg room service',
                'Baggage handling', 'Checkin service', 'Inflight service',
                'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'Generation'],
    'Importance': feature_importance_val
})

# Sort the DataFrame by importance in descending order
feature_importance_val_df = feature_importance_val_df.sort_values(by='Importance', ascending=False)

# Optionally, you can plot a bar chart for better visualization
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_val_df, palette='viridis')
plt.title('Gradient Boosting Feature Importance on Validation Set')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 범위 설정
neighbors_range = range(5, 30)

# 정확도 기록
accuracy_scores = []

for n_neighbors in neighbors_range:
    # KNN 모델 생성
    knn_model = KNeighborsClassifier(n_neighbors=n_neighbors)
    
    # 모델 학습
    knn_model.fit(X_train, y_train)
    
    # 검증 데이터로 예측
    y_val_pred_knn = knn_model.predict(X_val)
    
    # 정확도 계산 및 기록
    accuracy = accuracy_score(y_val, y_val_pred_knn)
    accuracy_scores.append(accuracy)

# 최적의 n_neighbors 선택
optimal_n_neighbors = neighbors_range[accuracy_scores.index(max(accuracy_scores))]

print(f'Optimal n_neighbors: {optimal_n_neighbors}')

# 선택된 최적의 n_neighbors로 KNN 모델 생성 및 테스트 데이터 평가
best_knn_model = KNeighborsClassifier(n_neighbors=optimal_n_neighbors)
best_knn_model.fit(X_train, y_train)

y_test_pred_best_knn = best_knn_model.predict(X_test)
test_accuracy_best_knn = accuracy_score(y_test, y_test_pred_best_knn)

print(f'Best KNN Test Accuracy: {test_accuracy_best_knn}')
```

```python
test_conf_matrix_knn = confusion_matrix(y_test, y_test_pred_best_knn)
conf_matrix_normalized = test_conf_matrix_knn / test_conf_matrix_knn.sum(axis=1)[:, np.newaxis]

# seaborn을 사용하여 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_normalized, annot=True, fmt=".4f", cmap="Blues", xticklabels=['Not Satisfied', 'Satisfied'], yticklabels=['Not Satisfied', 'Satisfied'])
plt.title('Normalized Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
