```python
import pandas as pd 
import seaborn as sns 
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

basket = pd.read_csv("Market_Basket_Optimisation.csv", header = None)
basket.head()
```

```python
records = []
for i in range (0, 7501):
    records.append([str(basket.values[i,j]) for j in range(0, 20)])
```

```python
TE = TransactionEncoder()
array = TE.fit(records).transform(records)

transf_df = pd.DataFrame(array, columns = TE.columns_)
transf_df
```

```python
basket_clean = transf_df.drop(['nan'], axis = 1)
basket_clean
```

```python
frequent_itemsets = apriori(basket_clean, min_support = 0.04, use_colnames = True)
frequent_itemsets.head()
```

```python
pattern_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
pattern_rules
```

```python
from mlxtend.frequent_patterns import fpgrowth

encode = TransactionEncoder()

encoded_array = encode.fit(basket_clean).transform(basket_clean)
encoded_array
```

```python
dataframe = pd.DataFrame(encoded_array, columns=encode.columns_)
dataframe
```
