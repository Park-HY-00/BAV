```python
import pandas as pd
import numpy as np
import seaborn as sns
print(pd.__version__)

titanic_df = pd.read_csv('titanic.csv')
```

```python
titanic_df.info()
```

```python
titanic_df.drop(columns='Name', inplace=True)
titanic_df.head(5)
```

```python
titanic_df.drop(columns=['Cabin', 'Embarked', 'Ticket'], inplace=True)
titanic_df.head(5)
```

```python
titanic_df.tail(10)
```

```python
titanic_df[titanic_df.duplicated()]
```

```python
titanic_df.drop_duplicates(inplace=True)
titanic_df[titanic_df.duplicated()]
titanic_df.tail(10)
```

```python
titanic_df.isnull().head(10)
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df.isnull().sum()
```

```python
titanic_df.head(10)
```

```python
titanic_df["Age"].interpolate(method = "linear", inplace = True)
titanic_df.head(10)
```

```python
titanic_df.dropna(inplace=True)
titanic_df.isnull().sum()
```

```python
titanic_df.columns
```

```python
titanic_df.rename(columns={
        'SibSp': 'Sibling.Spouse', 
}, inplace=True)

titanic_df.columns
```

```python
titanic_df.head(10)
```

```python
titanic_df.head(10)
titanic_df.replace({'Gender': {'male':0}}, inplace=True)
titanic_df.replace({'Gender': {'female': 1}}, inplace=True)
```

```python
sns.boxplot(y=titanic_df['Fare']);
```

```python
titanic_df.drop(titanic_df[titanic_df.Fare >= 500].index, inplace=True)
sns.boxplot(y=titanic_df['Fare']);
```

```python
titanic_df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic_df[['Fare']] = scaler.fit_transform(titanic_df[['Fare']])
titanic_df.describe()
```

```python
sns.heatmap(titanic_df.corr(), annot=True)
```
