```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy

print(sns.__version__)
```

```python
df = sns.load_dataset('taxis')
df.info()
```

```python
df.head()
```

```python
#plotting bar chart
my_pal = {"yellow": "y", "green": "g"}
sns.barplot(x='color', y='distance', data=df, errorbar=None, palette=my_pal, width=0.4)
plt.title("Fare prices by Taxi color")
plt.show()
```

```python
my_pal2 = {"credit card": "r", "cash": "b"}
sns.scatterplot(x='fare', y = 'tip', data = df, hue = 'payment', palette=my_pal2)
```

```python
sns.boxplot(x = 'color', y='total', data=df, width=0.5, palette=my_pal)
```

```python
sns.violinplot(x = 'passengers', y='tip', data=df, palette=my_pal, hue='color')
```

```python
sns.jointplot(x='fare', y='tip', data=df, hue='passengers', kind='kde')
```

```python
sns.lmplot(x="distance", y="tip", hue='color', data=df, palette=my_pal)
```
