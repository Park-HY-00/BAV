```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy

print(sns.__version__)
```

```python
df = sns.load_dataset('taxis')
df.info()
```

```python
df.head()
```

```python
#plotting bar chart
my_pal = {"yellow": "y", "green": "g"}
sns.barplot(x='color', y='distance', data=df, errorbar=None, palette=my_pal, width=0.4)
plt.title("distance by Taxi color")
plt.show()
```

```python
my_pal2 = {"credit card": "r", "cash": "b"}
sns.scatterplot(x='fare', y = 'tip', data = df, palette=my_pal2)
# hue  
```

```python
sns.boxplot(x = 'color', y='total', data=df, width=0.5, palette=my_pal)
```

```python
sns.violinplot(x = 'passengers', y='tip', data=df, palette=my_pal, hue='color')
```

```python
sns.lmplot(x="distance", y="tip", hue='color', data=df, palette=my_pal)
```

```python
sns.jointplot(x='distance', y='total', hue='passengers', data=df,kind="kde")
```

```python
sns.jointplot(x='distance', y='total', hue='passengers', data=df)
```

```python
sns.pairplot(df, hue="payment")
```

```python
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df,
    x="distance", y="tolls",
    palette=cmap, sizes=(10, 200),
)
# g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
# 'pickup_zone'     
unique_pickup_zones = df['pickup_zone'].unique()

#  
print(unique_pickup_zones)
```

```python
# 'pickup_zone'    
pickup_zone_counts = df['pickup_zone'].value_counts()

#  10   
pickup_zone_counts_over_150 = pickup_zone_counts[pickup_zone_counts >= 150]

#  
print(pickup_zone_counts_over_150)
```

```python
#  100  pickup_zone   DataFrame 
df_filtered = df[df['pickup_zone'].isin(pickup_zone_counts_over_150.index)]

# Seaborn  relplot 
sns.relplot(x='distance', y='total', hue='pickup_zone', data=df_filtered)

#   
plt.title('Distance vs Total with Pickup Zone (Counts >= 150)')

#  
plt.show()
```

```python
import seaborn.objects as so
# 'pickup_zone'    
pickup_zone_counts = df['pickup_zone'].value_counts()

#  10   
pickup_zone_counts_over_180 = pickup_zone_counts[pickup_zone_counts >= 180]

#  100  pickup_zone   DataFrame 
df_filtered2 = df[df['pickup_zone'].isin(pickup_zone_counts_over_180.index)]

(
    so.Plot(df_filtered2, y="distance", color="dropoff_borough")
    .pair(x=["pickup_zone", "pickup_borough"])
    .add(so.Dots())
)
```
